```python
def get_current_price(symbol):
    ticker = client.get_symbol_ticker(symbol=symbol)
    return float(ticker['price'])

# Call and print
price = get_current_price("BTCUSDT")
print("Current Price:", price)


# def get_current_price(symbol):
#     ticker = client.get_symbol_ticker(symbol=symbol)
#     return float(ticker['price'])
# get_current_price = (symbol)    
```

```python
def place_buy_order(symbol,trade_quantity):
    order = client.order_market_buy(symbol=symbol, quantity=trade_quantity)

    print(f"Buy order done:{order}")
place_buy_order(symbol,trade_qunatity)    
```

```python
def place_sell_order(symbol,trade_quantity):
    order = client.order_market_sell(symbol=symbol, quantity=trade_quantity)

    print(f"sell order done:{order}")
place_sell_order(symbol,trade_qunatity)    
```

```python
def trading_bot():
    in_position = False

    while True:
        current_price = get_current_price(symbol)
        print(f"Current price of {symbol}:{current_price}")
        if not in_position:
            if current_price < buy_price_threshold:
               print(f"Price is beolow {buy_price_threshold}.Placing buy order.")
               place_buy_order(symbol,trade_qunatity) 
               in_position = True
        else:
            if current_price > sell_price_threshold:
               print(f"Price is above {sell_price_threshold}.Placing sell order.")
               place_sell_order(symbol,trade_qunatity) 
               in_position = False
        time.sleep(3)        
               
```

```python
if __name__ == "__main__":
     trading_bot()
```
